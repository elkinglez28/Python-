```python
#iMAGEN BORROSA *
#Para hacer una imagen borrosa, puede usar el método GaussianBlur() de OpenCV.
# El GaussianBlur() utiliza el núcleo Gaussiano. La altura y el ancho del kernel deben ser un número positivo e impar.
# Luego hay que especificar la dirección X e Y 


img = cv2.imread("C:/Users/elkin/Pictures/ggg/programa/imagenes/IMAGEN4.png")

blur_image = cv2.GaussianBlur(img, (7,7), 0)

cv2.imshow('Original Image', img)

cv2.imshow('Blur Image', blur_image)

cv2.waitKey(0)

#En el recorte anterior, la imagen real se pasa a GaussianBlur() 
#junto con la altura y el ancho del núcleo y las direcciones X e Y.
```

```python
#Cambiar el tamaño de una imagen *
# Para redimensionar una imagen, puede usar el método resize() de openCV. En el método resize, puede especificar los 
#valores de los ejes x y y o el número de filas y columnas que indica el tamaño de la imagen.

import cv2

img = cv2.imread("C:/Users/elkin/Pictures/ggg/programa/imagenes/IMAGEN5.png")

newImg = cv2.resize(img, (0,0), fx=1.0, fy=1.0)

cv2.imshow('Resized Image', newImg)

cv2.waitKey(0)
```

```python
#RECORTAR IMAGENES *
#importar el módulo cv2 y leer la imagen y extraer el ancho y el alto de la misma
import cv2

img = cv2.imread("C:/Users/elkin/Pictures/ggg/programa/imagenes/IMAGEN3.png")

height, width = img.shape[0:2]

startRow = int(height*.10)

startCol = int(width*.15)

endRow = int(height*.50)

endCol = int(width*.85)

croppedImage = img[startRow:endRow, startCol:endCol]

cv2.imshow('Imagen Original', img)

cv2.imshow('Cortada', croppedImage)

cv2.waitKey(0)
```

```python
#Histogramas OpenCV Python
#el nivel de intensidad está en el rango de 0 a 255, en donde el valor 0 representa los color negro y 255 el color blanco

img = cv2.imread('C:/Users/elkin/Desktop/rr.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('C:/Users/elkin/Desktop/rr.jpg', img)

hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist, color='gray' )



plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()

cv2.destroyAllWindows()
```

```python
#Calcular histograma de imagen a color

img = cv2.imread('C:/Users/elkin/Desktop/dd.jpg')
cv2.imshow('C:/Users/elkin/Desktop/dd.jpg', img)

color = ('b','g','r')

for i, c in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0,256])

plt.show()

cv2.destroyAllWindows()
```
